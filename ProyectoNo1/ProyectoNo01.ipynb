```python
# Importar la Libreria Pandas
import pandas as pd5

# Crear un DataFrame vacío
df = pd.DataFrame(columns=['Dia', 'Cantidad', 'Precio'])

# Utilizar un Ciclo for para agregar datos de los dias de venta
for i in range(5):
    dato0 = 'Dia ' + str(i + 1)
    dato1 = float(input(f'Ingrese la Cantidad del Producto {i+1}: '))
    dato2 = float(input(f'Ingrese el Precio de Producto {i+1}: '))
    df.loc[i] = [dato0, dato1, dato2]

# Calcular Venta Sin Descuento
df['VentaSinDesc'] = df['Cantidad'] * df['Precio']

# Mostrar los Datos del DataFrame
print(df)

# Calcular el Descuento 5%
df['Descuento'] = df['Cantidad'] * df['Precio'] * 0.05

# Calcular Venta Con Descuento
df['VentaConDesc'] = df['VentaSinDesc'] - df['Descuento']

# Establecer la Columna 'Dia' como el índice
df = df.set_index('Dia')

# Mostrar los datos del DataFrame
print(df)
```

```python
# Importar la Libreria Pandas
import pandas as pd

# Definir la funcion que calcule los datos estadisticos
def calcular_estadisticos(ingresos):
    ingresos_serie = pd.Series(ingresos)
    minimo = ingresos_serie.min()
    maximo = ingresos_serie.max()
    media = ingresos_serie.mean()
    estadisticos = pd.Series([minimo, maximo, media], index=['Mínimo', 'Máximo', 'Media'])
    return estadisticos

# Definir el diccionario con la venta de cada cliente
ingresos_mensuales = {
    'Cliente 1': 1011,
    'Cliente 2': 1600,
    'Cliente 3': 2500,
    'Cliente 4': 1100,
    'Cliente 5': 1900
}

# Mostrar los datos del Diccionario
print(ingresos_mensuales)

# Calcular los datos estaditicos invocando la funcion
serie = calcular_estadisticos(ingresos_mensuales)

# Mostrar los datos estadisticos calculados a partir de los datos del diccionario
print(serie)
```

**Ejercicio N°3:**

```python
# Importar la Libreria Pandas
import pandas as pd

# Crear un DataFrame con los siguiente datos de cada mes
datos = {
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo'],
    'Ventas': [30500, 35600, 28300, 33900, 42500],
    'Gastos': [22000, 23450, 18100, 35700, 32450]
}
df = pd.DataFrame(datos)

# Mostrar los datos del DataFrame
print(df)

# Crear un nueva columna a partir de la siguiente operacion Gasto/Ventas*100
df['Columna 1'] = (df['Gastos'] / df['Ventas']) * 100

# Mostrar los datos del DataFrame
print(df)

# Crear un nueva columna con la siguiente condicion Si Ventas>= 30000, imprimir “Meta Superada”,caso contrario imprimir “Meta no Superada”
df['Columna 2'] = df['Ventas'].apply(lambda x: 'Meta Superada' if x >= 30000 else 'Meta no Superada')

# Mostrar los datos del DataFrame
print(df)

# Crear una nueva columna con la siguiente condicion Si Gasto>Ventas, calcular el sobre gasto
df['Columna 3'] = df.apply(lambda row: row['Gastos'] - row['Ventas'] if row['Gastos'] > row['Ventas'] else 0, axis=1)

# Mostrar los datos del DataFrame
print(df)

# Obtener el valor de los ultimos 3 meses
ultimos_3_meses = df['Ventas'].tail(3)

# Obtener el total de las ventas de los ultimos 3 meses
suma_ultimos_3_meses = ultimos_3_meses.sum()

#Mostrar el total de las ventas de los ultimos 3 meses
print('Suma de los últimos 3 Meses:', suma_ultimos_3_meses)

# Filtrar los gastos de febrero a abril
gastos_feb_abr = df[df['Mes'].isin(['Febrero', 'Marzo', 'Abril'])]['Gastos']

# Calcular el total de los gastos
total_gastos = gastos_feb_abr.sum()

# Mostrar el total de gastos de febrero a abril
print("Total de gastos de febrero a abril:", total_gastos)
```
